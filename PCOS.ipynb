```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pcos_df = pd.read_csv(r"C:\Users\roxy3\OneDrive\UNCC\Spring 2025\Data mining\archive\PCOS_data_without_infertility.csv")
```

```python
pcos_df.head()
```

```python
pcos_df.columns = pcos_df.columns.str.strip()
```

```python
pcos_df.head()
```

```python
pcos_df['BMI'] = pd.to_numeric(pcos_df['BMI'], errors='coerce')
```

```python
pcos_df.loc[:, 'BMI'] = pcos_df['BMI'].fillna(pcos_df['BMI'].mean())
```

```python
print(pcos_df['BMI'].isnull().sum())
```

```python
print(pcos_df.head())
```

```python
pcos_df.drop(columns=['Sl. No'], inplace=True, errors='ignore')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(pcos_df.corr(), cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Clinical Features")
plt.show()
```

```python
plt.figure(figsize=(14, 10))  
sns.heatmap(pcos_df.corr(), cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5)
plt.xticks(rotation=90)  
plt.yticks(rotation=0) 
plt.title("Correlation Heatmap of Clinical Features", fontsize=16)
plt.show()

```

```python
sns.heatmap(pcos_df.corr(), cmap="RdBu_r", annot=True, fmt=".2f", linewidths=0.5)
```

```python
plt.figure(figsize=(16, 12))  
sns.heatmap(pcos_df.corr(), cmap="RdBu_r", annot=True, fmt=".1f", linewidths=0.5, annot_kws={"size": 8})
plt.xticks(rotation=90, fontsize=10)  
plt.yticks(fontsize=10)
plt.title("Correlation Heatmap of Clinical Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(14, 10)) 
sns.heatmap(
    pcos_df.corr(), 
    cmap="coolwarm", 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    annot_kws={"size": 8}  
)
plt.xticks(rotation=90, fontsize=10)  
plt.yticks(fontsize=10)
plt.title("Correlation Heatmap of Clinical Features", fontsize=16)
plt.show()
```

```python
corr_matrix = pcos_df.corr()
strong_corr = corr_matrix[(corr_matrix >= 0.5) | (corr_matrix <= -0.5)]
sns.heatmap(strong_corr, cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5)
```

```python
plt.figure(figsize=(12, 10))  
sns.heatmap(
    strong_corr, 
    cmap="coolwarm", 
    annot=True, 
    fmt=".2f", 
    linewidths=0.5, 
    annot_kws={"size": 10}  
)
plt.xticks(rotation=45, ha="right", fontsize=10)  
plt.yticks(fontsize=10)
plt.title("Strong Correlations in Clinical Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="PCOS (Y/N)", y="BMI", data=pcos_df, inner="quartile")
plt.title("BMI Distribution Among PCOS and Non-PCOS Individuals")
plt.xlabel("PCOS Diagnosis")
plt.ylabel("BMI")
plt.xticks(ticks=[0, 1], labels=["Non-PCOS", "PCOS"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pcos_df, x="Age (yrs)", y="BMI", hue="PCOS (Y/N)", alpha=0.7)
plt.title("BMI vs. Age with PCOS Diagnosis")
plt.xlabel("Age (years)")
plt.ylabel("BMI")
plt.legend(["Non-PCOS", "PCOS"])
plt.show()
```

```python
plt.figure(figsize=(12, 6))

#LH levels
plt.subplot(1, 3, 1)
sns.boxplot(x="PCOS (Y/N)", y="LH(mIU/mL)", data=pcos_df)
plt.title("LH Levels in PCOS vs. Non-PCOS")
plt.xlabel("PCOS Diagnosis")
plt.ylabel("LH (mIU/mL)")
plt.xticks([0, 1], ["Non-PCOS", "PCOS"])

# FSH levels
plt.subplot(1, 3, 2)
sns.boxplot(x="PCOS (Y/N)", y="FSH(mIU/mL)", data=pcos_df)
plt.title("FSH Levels in PCOS vs. Non-PCOS")
plt.xlabel("PCOS Diagnosis")
plt.ylabel("FSH (mIU/mL)")
plt.xticks([0, 1], ["Non-PCOS", "PCOS"])

# Testosterone levels
plt.subplot(1, 3, 3)
sns.boxplot(x="PCOS (Y/N)", y="Testosterone(ng/dL)", data=pcos_df)
plt.title("Testosterone Levels in PCOS vs. Non-PCOS")
plt.xlabel("PCOS Diagnosis")
plt.ylabel("Testosterone (ng/dL)")
plt.xticks([0, 1], ["Non-PCOS", "PCOS"])

```
